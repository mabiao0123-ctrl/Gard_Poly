```python
# Construct two random decreasing sequences of negative numbers with seq2[i] >= seq1[i]
import numpy as np
import sympy as sp
import Garding as gd
n = 6 # Length of sequences

# First sequence: random decreasing negative numbers
seq1 = -np.sort(np.abs(np.random.uniform(0.1, 13, n)))

# Second sequence: for each i, pick a random negative number >= seq1[i]
seq2 = np.array([np.random.uniform(seq1[i], seq1[i]*np.random.uniform(0,1)) for i in range(n)])
seq2 = np.sort(seq2)[::-1]  # Ensure decreasing order

print('Sequence 1:', seq1)
print('Sequence 2:', seq2)
print('Check: seq2[i] >= seq1[i] for all i:', np.all(seq2 >= seq1))
t=sp.symbols('t')
f1=sp.Poly(gd.constr_f(list(seq1)), t)
f2=sp.Poly(gd.constr_f(list(seq2)), t)
print('f1:', f1)
def domination(f1, f2):
    # Check if f2 is in the proper position with respect to f1
    # This is a placeholder for the actual implementation
    if isinstance(f1, list) or isinstance(f2, list):
        coeffs1 = f1
        coeffs2 = f2
        f1=sp.Poly(coeffs1, t)
    else:
        if isinstance(f1, np.ndarray) or isinstance(f2, np.ndarray):
            coeffs1 = f1
            coeffs2 = f2
            f1=sp.Poly(coeffs1, t)
        else:
            coeffs1 = f1.all_coeffs()
            coeffs2 = f2.all_coeffs()
    if gd.UpsilonStable(coeffs2)==False:
        return False
    
    #rootseq1 = gd.dervf(coeffs1)
    rootseq2 = gd.dervf(coeffs2)
    f1_poly = f1.as_expr() if hasattr(f1, 'as_expr') else sp.Poly(f1, t).as_expr()
    d = len(rootseq2)
    for k, x in enumerate(rootseq2):
        for j in range(d-k):
            deriv = sp.diff(f1_poly, t, j+k)
            val = deriv.subs(t, x)
            #print(f"{k}-th derivative of f1 at rootseq2[{k}] = {val}")
            if val < 0:
                return False
    return True

def proper_position(f1, f2):
    # Check if f2 is in proper position with respect to f1
    if isinstance(f1, list) or isinstance(f2, list):
        coeffs1 = f1
        coeffs2 = f2
        f1=sp.Poly(coeffs1, t)
        f2=sp.Poly(coeffs2, t)
    else:
        if isinstance(f1, np.ndarray) or isinstance(f2, np.ndarray):
            coeffs1 = f1
            coeffs2 = f2
            f1=sp.Poly(coeffs1, t)
            f2=sp.Poly(coeffs2, t)
        else:
            coeffs1 = f1.all_coeffs()
            coeffs2 = f2.all_coeffs()
    if abs(len(coeffs2)-len(coeffs1))>=2:
        return False
    if gd.UpsilonStable(coeffs2)==False or gd.UpsilonStable(coeffs1)==False:
        return False
    
    rootseq1 = gd.dervf(coeffs1)
    rootseq2 = gd.dervf(coeffs2)
    if len(rootseq1)==len(rootseq2): 
        if np.all(np.array(rootseq2) >= np.array(rootseq1))==False:
            return False
    d = len(rootseq2)
    for k, x in enumerate(rootseq1):
        deriv = sp.diff(f2.as_expr(), t, k)
        val = deriv.subs(t, x)
        #print(f"{k}-th derivative of f2 at rootseq1[{k}] = {val}")
        if val > 0:
            return False
    return True
proper_position(f1, f2)
```
